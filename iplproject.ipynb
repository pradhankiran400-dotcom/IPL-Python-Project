```python
import pandas as pd
import numpy as np

class IPLAnalyzer:

    def __init__(self, csv_path="deliveries.csv"):
        self.df = pd.read_csv(csv_path)
        self.menu()

    def player_data(self, player):
        return self.df[self.df['batsman'] == player]

    def short_against_team(self, player, short_type, opp):

        p_data = self.df['batsman'] == player
        opp_data = self.df['bowling_team'] == opp
        six = self.df['batsman_runs'] == 6
        four = self.df['batsman_runs'] == 4

        if short_type == 6:
            count = self.df[p_data & opp_data & six]['batsman_runs'].count()
            print(f"{player} hit {count} sixes against {opp}")
        elif short_type == 4:
            count = self.df[p_data & opp_data & four]['batsman_runs'].count()
            print(f"{player} hit {count} fours against {opp}")
        else:
            print("enter valid short_type (4 or 6)")

    def teams_run(self, team):
        t = self.df[self.df['batting_team'] == team]
        bat = t.groupby('batsman')
        result = bat['batsman_runs'].sum().sort_values(ascending=False).head(20)
        return result


    def menu(self):
        while True:
            choice = input('''
        1. Team-wise each batsman total runs
        2. Batsman run (4 or 6) against a team
        3. Individual player full data
        0. Exit

        Enter your choice: ''')

            if choice == '1':
                team = input("Enter Team Name: ")
                result = self.teams_run(team)
                print(result)

            elif choice == '2':
                player = input("Enter Player Name: ")
                opp = input("Enter Opponent Team: ")
                short = input("Enter shot type (4 or 6): ")
                if short in ['4', '6']:
                    self.short_against_team(player, int(short), opp)
                else:
                    print("Invalid shot type, enter 4 or 6")

            elif choice == '3':
                player = input("Enter Player Name: ")
                print(self.player_data(player))

            elif choice == '0':
                print("Exiting..")
                break

            else:
                print(" Invalid Choice, try again.")
```
